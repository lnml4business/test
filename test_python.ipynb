```python
from datetime import datetime

def clear_screen(x):
    print("\n" * x)
    print('screen cleared')

def get_data():
    data_path = input("Welcome!!! Please enter the file path:")
    file = open(data_path, 'r')
    date_list = []
    close_list = []
    for line in file.readlines():
        fnames = line.rstrip().split(',') 
        b = 0
        for fname in fnames:
            if b == 0:
                date_list.append(fname)
            else:
                close_list.append(fname)
            b += 1
    return date_list[1:], close_list[1:]


def time(input_data):
    time_list = []
    for i in input_data[0]:
        time_value = datetime(year=int('19' + i[0:2]), month=int(i[2:4]), day=int(i[4:6]))
        time_value = time_value.strftime('%Y/%m/%d')
        time_list.append(time_value)
    return time_list


def median(input_data):
    median_list = []
    input_data_list = input_data[1]
    for i in range(len(input_data_list)):
        data_list = input_data_list[:i+1]
        size = len(data_list)
        midPos = size // 2
        if size == 1: 
            median = 0
        else:
            if size % 2 == 0:
                median = (float(input_data_list[midPos]) + float(input_data_list[midPos-1])) / 2
            else:
                median = float(input_data_list[midPos])
        median_list.append(median)
    return median_list

def mean(input_data):
    mean_list = []
    input_data_list = input_data[1]
    for i in range(len(input_data_list)):
        data_list = input_data_list[:i+1]
        sum = 0.0
        size = len(data_list)
        if size == 1:
            mean = 0
            mean_list.append(mean)
        else:
            for num in data_list:
                sum = sum + float(num)
            mean = sum / len(data_list)
            mean_list.append(mean)
    return mean_list

def updown_gainloss(input_data):
    updown_list = []
    gainlost_list = []
    close_added_list = []
    close_added_list.append(input_data[1][0])
    for i in input_data[1]:
        close_added_list.append(i)
        
    zip_object = zip(input_data[1], close_added_list)
    for list1_i, list2_i in zip_object:
        updown_list.append(float(list1_i) - float(list2_i))
        gainlost_list.append((float(list1_i) - float(list2_i)) / float(list2_i) * 100)

    return updown_list, gainlost_list

def formated_list(my_list):
    formatted_list = []
    for item in my_list:
        formatted_list.append("%.2f"%item)
    return formatted_list

def main():
    data = get_data()
    time_list = time(data)
    close_list = data[1]
    mean_list = formated_list(mean(data))
    median_list = formated_list(median(data))
    updown_list = formated_list(updown_gainloss(data)[0])
    gainloss_list = formated_list(updown_gainloss(data)[1])

    with open('file.csv', 'w') as fh:
        cols = ['Date', 'Close', 'Mean', 'Median', 'Up/Down', 'Gain/Loss']
        fh.write('\t'.join(cols) + '\n')

        # here is where you unpack everything
        for row in zip(time_list, close_list, mean_list, median_list, updown_list, gainloss_list):
            s = " ".join(map(str, row))
            fh.write(s+'\n')

if __name__ == '__main__' :
    main()
```
